## <font color=#09829c> Fusionner les datasets</font>

Le tableau suivant, explique comment nous allons fusionner les datasets :<br>

|Left_df| Right_df|On_Key|Result_df|
|:-|:-|:-|:-|
|customers|orders|"customer_id"|merge_step_1|
|merge_step_1|geolocation["geolocation_lat", "geolocation_lng"]|"zip_code_prefix"|merge_step_2|
|merge_step_2|order_items<br>order_payments<br>order_reviews|"order_id"|merge_step_3|
|merge_step_3|sellers|"seller_id"|merge_step_4|
|merge_step_4|products|"product_id"|merge_step_5|
|merge_step_5|product_category_name_translation|"product_category_name"|merged_df|

```python
%%time

# Etape n°1 pour la fusion des datasets
merge_step_1 = pd.merge(customers, orders, on="customer_id")

# Etape n°2 pour la fusion des datasets
merge_step_2 = pd.merge(
    merge_step_1,
    geolocation.groupby(["geolocation_zip_code_prefix"])[
        "geolocation_lat", "geolocation_lng"
    ].mean(),
    left_on="customer_zip_code_prefix",
    right_on="geolocation_zip_code_prefix",
)

# Etape n°3 pour la fusion des datasets
merge_step_3 = reduce(
    lambda left, right: pd.merge(left, right, on=["order_id"]),
    [merge_step_2, order_items, order_payments, order_reviews],
)

# Etape n°4 pour la fusion des datasets
merge_step_4 = pd.merge(merge_step_3, sellers, on="seller_id")

# Etape n°5 pour la fusion des datasets
merge_step_5 = pd.merge(merge_step_4, products, on="product_id")

# Etape n°6 pour la fusion des datasets
merged_df = pd.merge(
    merge_step_5, product_category_name_translation, on="product_category_name"
)
```

- Création d'une fonction pour détecter et supprimer les outliers selon la méthode Tuekey.

```python
# Tukey's method
def tukeys_method(dataframe):
    """
    Cette fonction calcul le nombre d'outliers en utilisant la méthode de Tukey.
    Input : dataframe.
    Output: modified dataframe
    """
    # Liste des colonnes avec des outliers univariés
    numerical_columns = (
        dataframe.select_dtypes(include=["int64", "float64"])
        .drop(
            ["geolocation_lat", "geolocation_lng"],
            axis=1,
        )
        .columns.tolist()
    )

    for column in numerical_columns:

        Q1 = dataframe[column].quantile(0.25)
        Q3 = dataframe[column].quantile(0.75)
        IQR = Q3 - Q1

        # calcul des limites d'outer_fence
        outer_fence_low = Q1 - 3 * IQR
        outer_fence_high = Q3 + 3 * IQR

        # calcul du nombre d'outliers détectés
        outlier_detected_before = np.sum(dataframe[column] < outer_fence_low) + np.sum(
            dataframe[column] > outer_fence_high
        )

        # Afficher le nombre d'outliers détectés
        print(f"Colonne :{column}")
        print("Nbr d'outliers détectés AVANT imputation:", outlier_detected_before)

        # Remplacer les outliers par les valeurs de low fence et high fence.
        dataframe.loc[dataframe[column] < outer_fence_low, [column]] = outer_fence_low
        dataframe.loc[dataframe[column] > outer_fence_high, [column]] = outer_fence_high

        # calcul du nombre d'outliers détectés
        outlier_detected_after = np.sum(dataframe[column] < outer_fence_low) + np.sum(
            dataframe[column] > outer_fence_high
        )

        print("Nbr d'outliers détectés APRES imputation :", outlier_detected_after)

        print("-" * 80)

    return None
```

## <font color=#09829c> Encoder la feature `order_month` en utilisant la méthode **Cyclical Encoding**</font>

- La méthode cyclique consiste à transformer les données en deux dimensions à l'aide d'une transformation sinus et consin.


![alt text](https://miro.medium.com/max/343/1*70cevmU8wNggGJEdLam1lw.png "Title")


- xsin=sin(2∗π∗xmax(x)) 
- xcos=cos(2∗π∗xmax(x))

- Fonction pour **Cyclical Encoding**

```python
def cyclical_encoding(dataframe, month_column):

    dataframe["month_sin"] = np.sin(2 * np.pi * dataframe[month_column] / 12)
    dataframe["month_cos"] = np.cos(2 * np.pi * dataframe[month_column] / 12)

    dataframe.drop([month_column], axis=1, inplace=True)

    return dataframe
```

```python
def auto_clean(input_file_path, out_file_path_cleaned):

    """
    Cette fonction retourne un dataframe après avoir réalisé les opérations suivantes :
    - Etape 01 : Charger les 9 datasets.
    - Etape 02 : Fusionner les 9 datasets.
    - Etape 03 : Filtrer les colonnes du dataset.
    - Etape 04 : Dectection et Imputation des Outliers.
    - Etape 05 : Feature Engineering.
    - Etape 06 : Pre-processing des données".
    - Etape 07 : Sauvegarder le dataset cleaned & transformed.

    """
    #  Etape 01 : Charger les 9 datasets.
    print("-" * 100)
    print("Etape 01 : Charger les 9 datasets.")

    # Location des datasets
    data_set_location = glob(f"{input_file_path}/*.csv")

    # Charger les datasets
    data_set_names = []

    for name in data_set_location:
        data_frame = (
            name.split("\\")[1]
            .replace("-", "_")
            .split(".")[0]
            .replace("olist_", "")
            .replace("_dataset", "")
        )
        exec(f"{data_frame} = pd.read_csv('{name}')")
        data_set_names.append(data_frame)

    # Créer une liste des datasets
    dataframes_list = [
        customers,
        geolocation,
        orders,
        order_items,
        order_payments,
        order_reviews,
        products,
        sellers,
        product_category_name_translation,
    ]

    #  Etape 02 : Nettoyer les 9 datasets.
    print("-" * 100)
    print("Etape 02 : Fusionner les 9 datasets.")

    # step n°1 pour la fusion des datasets
    step_1 = pd.merge(customers, orders, on="customer_id")

    # step n°2 pour la fusion des datasets
    step_2 = pd.merge(
        step_1,
        geolocation.groupby(["geolocation_zip_code_prefix"])[
            "geolocation_lat", "geolocation_lng"
        ].mean(),
        left_on="customer_zip_code_prefix",
        right_on="geolocation_zip_code_prefix",
    )

    # step n°3 pour la fusion des datasets
    step_3 = reduce(
        lambda left, right: pd.merge(left, right, on=["order_id"]),
        [step_2, order_items, order_payments, order_reviews],
    )

    # step n°4 pour la fusion des datasets
    step_4 = pd.merge(step_3, sellers, on="seller_id")

    # step n°5 pour la fusion des datasets
    step_5 = pd.merge(step_4, products, on="product_id")

    # step n°6 pour la fusion des datasets
    merged_df = pd.merge(
        step_5, product_category_name_translation, on="product_category_name"
    )

    #  Etape 03 : Filtrer les colonnes du dataset..
    print("-" * 100)
    print("Etape 03 : Filtrer les colonnes du dataset.")

    # Filtre
    filtre_columns = [
        "customer_id",
        "order_id",
        "order_item_id",
        "price",
        "freight_value",
        "payment_installments",
        "payment_value",
        "customer_city",
        "customer_state",
        "geolocation_lat",
        "geolocation_lng",
        "review_score",
        "payment_type",
        "product_category_name_english",
        "order_delivered_customer_date",
        "order_purchase_timestamp",
    ]

    # Filtrer les colonnes
    merged_df = merged_df[filtre_columns].copy()

    for column in ["order_item_id", "customer_id", "order_id"]:
        merged_df[column] = merged_df[column].astype("object")

    #  Etape 04 : Dectection et Imputation des Outliers.
    print("-" * 100)
    print("Etape 04 : Dectection et Imputation des Outliers")

    tukeys_method(merged_df)

    #  Etape 05 : Nettoyer les 9 datasets.
    print("Etape 05 : Feature Engineering.")

    # Creation d'une nouvelle feature Delivery Time in day
    merged_df["Delivery_Lead_Time_in_days"] = (
        merged_df["order_delivered_customer_date"]
        - merged_df["order_purchase_timestamp"]
    ).dt.days

    # Creation d'une nouvelle feature Order Month
    merged_df["order_month"] = pd.Series(
        merged_df["order_purchase_timestamp"].values.astype("datetime64[M]")
    ).dt.month

    # Creation d'une nouvelle feature Month & Year
    merged_df["order_month_year"] = pd.to_datetime(
        merged_df["order_purchase_timestamp"]
    ).dt.to_period("M")

    # Creation des autres features

    df_reduced = (
        merged_df.pivot_table(
            index="customer_id",
            values=[
                "order_id",
                "order_item_id",
                "price",
                "freight_value",
                "payment_installments",
                "payment_value",
                "customer_city",
                "customer_state",
                "geolocation_lat",
                "geolocation_lng",
                "review_score",
                "payment_type",
                "product_category_name_english",
                "Delivery_Lead_Time_in_days",
                "order_month",
                "order_month_year",
            ],
            aggfunc={
                "order_id": pd.Series.nunique,  # nombre de commande par client
                "order_item_id": np.max,  # nombre de produits par commande par client
                "price": np.mean,  # prix moyen des produits achetés par client
                "freight_value": np.mean,  # frais moyen de livraison des produits achetés par client
                "payment_installments": np.mean,  # nombre moyen de paiement des achats par client
                "payment_value": np.mean,  # dépense totale moyenne par client (cout produit + frais de livraison)
                "customer_city": pd.Series.unique,  # Ville du client
                "customer_state": pd.Series.unique,  # Département du client
                "geolocation_lat": pd.Series.unique,
                "geolocation_lng": pd.Series.unique,
                "review_score": np.mean,
                "payment_type": calcul_mode,  # pd.Series.mode,#scipy.stats.mode,
                "product_category_name_english": calcul_mode,  # La categorie de produit la plus achetée par client
                "Delivery_Lead_Time_in_days": np.mean,  # nombre moyen de jour de livraison
                "order_month": "first",  # mois de la commande
                "order_month_year": "first",  # mois et année de la commande
            },
        )
        .sort_values(by="payment_value", ascending=False)
        .rename(
            columns={
                "order_id": "Nbr_order_placed_per_customer",
                "order_item_id": "Nbr_of_products_bought_per_order",
                "price": "average_product_price",
                "freight_value": "average_freight_value",
                "payment_installments": "payment_factions",
                "payment_value": "average_payement_value",
                "geolocation_lat": "latitude",
                "geolocation_lng": "longitude",
                "product_category_name_english": "most_bought_product_category",
            }
        )
    )

    # Ré-organiser les indexs du nouveau df
    df_reduced = df_reduced.reindex(sorted(df_reduced.columns, reverse=True), axis=1)

    # Supprimer les valeurs négatives dans la feature Delivery Time
    df_reduced = df_reduced[df_reduced["Delivery_Lead_Time_in_days"] > 0]

    # Supprimer les Delivery Time supérieur à 3 mois (90 jours)
    df_reduced = df_reduced[df_reduced["Delivery_Lead_Time_in_days"] < 90]

    #  Etape 06 : Pre-processing des données".
    print("-" * 100)
    print("Etape 06 : Pre-processing des données")

    # Création de nouvelles catégories de produits
    product_category_dict = {
        # Appliances & Housewares
        "home_appliances": "Appliances & Housewares",
        "home_appliances_2": "Appliances & Housewares",
        "small_appliances": "Appliances & Housewares",
        "small_appliances_home_oven_and_coffee": "Appliances & Housewares",
        "la_cuisine": "Appliances & Housewares",
        "housewares": "Appliances & Housewares",
        "kitchen_dining_laundry_garden_furniture": "Appliances & Housewares",
        # Construction
        "construction_tools_construction": "construction",
        "construction_tools_lights": "construction",
        "construction_tools_safety": "construction",
        "costruction_tools_garden": "construction",
        "costruction_tools_tools": "construction",
        "home_construction": "construction",
        "air_conditioning": "construction",
        # Fashion
        "fashio_female_clothing": "fashion",
        "fashion_bags_accessories": "fashion",
        "fashion_childrens_clothes": "fashion",
        "fashion_male_clothing": "fashion",
        "fashion_shoes": "fashion",
        "fashion_sport": "fashion",
        "fashion_underwear_beach": "fashion",
        "watches_gifts": "fashion",
        # Electronics
        "electronics": "electronics",
        "tablets_printing_image": "electronics",
        "telephony": "electronics",
        "audio": "electronics",
        "fixed_telephony": "electronics",
        "dvds_blu_ray": "electronics",
        "computers": "electronics",
        "computers_accessories": "electronics",
        "consoles_games": "electronics",
        "cds_dvds_musicals": "electronics",
        "musical_instruments": "electronics",
        # Food
        "food": "food",
        "food_drink": "food",
        "drinks": "food",
        # Health - Beauty & Hygiene
        "diapers_and_hygiene": "Health - Beauty & Hygiene",
        "perfumery": "Health - Beauty & Hygiene",
        "health_beauty": "Health - Beauty & Hygiene",
        "baby": "Health - Beauty & Hygiene",
        # Art - Culture
        "art": "Art - Culture",
        "arts_and_craftmanship": "Art - Culture",
        "books_general_interest": "Art - Culture",
        "books_imported": "Art - Culture",
        "books_technical": "Art - Culture",
        "music": "Art - Culture",
        "cine_photo": "Art - Culture",
        # Home_furniture_supplies
        "furniture_bedroom": "Home_furniture_supplies",
        "furniture_decor": "Home_furniture_supplies",
        "furniture_living_room": "Home_furniture_supplies",
        "furniture_mattress_and_upholstery": "Home_furniture_supplies",
        "christmas_supplies": "Home_furniture_supplies",
        "bed_bath_table": "Home_furniture_supplies",
        "home_comfort_2": "Home_furniture_supplies",
        "home_confort": "Home_furniture_supplies",
        "party_supplies": "Home_furniture_supplies",
        # Garden
        "flowers": "Garden",
        "garden_tools": "Garden",
        "pet_shop": "Garden",
        # Industry_commerce
        "agro_industry_and_commerce": "Industry_commerce",
        "industry_commerce_and_business": "Industry_commerce",
        "market_place": "Industry_commerce",
        # Office_furniture
        "office_furniture": "Office_furniture",
        "stationery": "Office_furniture",
        # Security
        "security_and_services": "Security",
        "signaling_and_security": "Security",
        # auto
        "auto": "Auto",
        # Sport
        "sports_leisure": "Sports",
        "luggage_accessories": "Sports",
        # toys
        "toys": "toys",
        # Other
        "cool_stuff": "cool_stuff",
    }

    # Grouper les catégories de produits
    df_reduced["most_bought_product_category"].replace(
        product_category_dict, inplace=True
    )

    # Passer les features "customer_city" et "customer_state" en index pour réduire la complexité du One Hot Encoding
    df_reduced.set_index(
        ["customer_city", "customer_state", "order_month_year"], inplace=True
    )

    # Pre-processing des données (One Hot Encoding des variables Qualitative + StandScaler des variables Quantitatives)

    # Encoder la order_month
    df_reduced = cyclical_encoding(dataframe=df_reduced, month_column="order_month")

    # Liste des variables Quantitatives
    numerical_features = (
        df_reduced.select_dtypes(include=["int64", "float64"]).columns.tolist()
        #         .drop(["month_sin", "month_cos"], axis=1)
        #         .columns.tolist()
    )

    # Liste des variables Qualitatives
    categorical_features = df_reduced.select_dtypes(
        exclude=["int64", "float64"]
    ).columns.tolist()

    # Créer un Transformer pour appliquer la transformation des données avec MinMaxScaler()
    MinMax_Transformer = ColumnTransformer(
        transformers=[
            (
                "onehot",
                OneHotEncoder(sparse=False, handle_unknown="ignore"),
                categorical_features,
            ),
            ("scaler", MinMaxScaler(), numerical_features),
        ],
        remainder="passthrough",
    )

    # Transformation des données
    MinMax_Scaled_data = MinMax_Transformer.fit_transform(df_reduced)

    # Créer un nouveau dataframe avec les données transformées
    transformed_data_column = []
    transformed_data_columns = (
        MinMax_Transformer.named_transformers_["onehot"]
        .get_feature_names()
        .tolist()
        .copy()
    )
    transformed_data_columns.extend(numerical_features)
    #     transformed_data_columns.extend(["month_sin", "month_cos"])

    MinMax_Scaled_df = pd.DataFrame(
        MinMax_Scaled_data, index=df_reduced.index, columns=transformed_data_columns
    )

    #  Etape 07 : Sauvegarder le dataset cleaned & transformed".
    print("-" * 100)
    print("Etape 07 : Sauvegarder le dataset cleaned & transformed.")

    # Sauvgarder le dataframe nettoyé avant preprocessing
    df_reduced.to_csv(f"{out_file_path_cleaned}cleaned_data.csv")

    # Sauvgarder le dataframe nettoyé et preprocessed avec MinMaxScaler()
    MinMax_Scaled_df.to_csv(f"{out_file_path_cleaned}MinMax_Scaled_Data.csv")

    print(
        f"Adresse complete de fichier nettoyé {out_file_path_cleaned}cleaned_data.csv"
    )

    print(
        f"Adresse complete de fichier nettoyé et pre-processed avec MinMaxScaler(){out_file_path_cleaned}MinMax_Scaled_Data.csv.csv"
    )

    print("-" * 100)

    return print(
        "Toutes les étapes du nettoyage du dataset ont été réalisé avec succèss "
    )
```
